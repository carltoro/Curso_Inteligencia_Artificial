### Iterando sobre los elementos de una lista

Para esto usaremos el ciclo for, cuya declaración se puede ver en la siguiente imagen:

![Image in a markdown cell]( https://www.learnbyexample.org/wp-content/uploads/python/Python-for-Loop-Syntax.png)

[fuente](https://www.learnbyexample.org/python-for-loop/)

En python, el ciclo for itera sobre estructuras iterables como las listas, tomando los elementos en el orden que aparecen en la variable iterable.

Notar que el uso de else: es opcional

````python
# Ejemplos:
colores = ['rojo', 'verde', 'azul', 'amarillo']
for x in colores:
  print(x)
````

Para iterar sobre un conjunto de instrucciones un número determinado de veces,
se puede usar la función de python ``range()``, esta función genera una secuencia de números desde cero hasta (sin incluir) un número especificado, se declara como: ``range(start,stop,step)``

````python
# Ejemplo
for x in range(7):
    print(x)
````

````python
#Introduciendo condiciones
# Ejemplo
for num in range(10):
    if (num%2)==0:
      print(f"{num} es par")
    else:
      print(f"{num} es impar")
````

Otro aspecto interesante, es si quisieramos ocupar los indices de los elementos de alguna variable iterable, con la función ``enumerate()`` podremos obtener tanto el índice como el elemento para usarlos en el ciclo.

````python
one_piece = ['Monkey', 'D.', 'Luffy']

for index, value in enumerate(one_piece):
    print(index, value)
````

Por otro lado, si quiseramos iterar sobre más de una lista a la vez podemos ocupar la función ``zip()``, por ejemplo:

````python
nombre = ['Bob', 'Sam', 'Max']
edad   = [25, 35, 30]
for x, y in zip(nombre, edad):
    print(x,":", y)
````

Finalmente, una forma de crear listas simplificando el código al usar ciclos for es con las ``list comprehensions``, esta sintaxis se puede escribir de la siguiente forma:

````python
# Ejemplo al crear una lista con los cuadrados de números entre 0 y 9
cuadrados = [num**2 for num in range(10)]
print(cuadrados)
````

````python
# Ejemplo al crear una lista con los cudrados de los números pares entre 0 y 9
cuadrados_pares = [num**2 for num in range(10) if num % 2 == 0]
print(cuadrados_pares)
````

En la celda anterior se uso el ```operador %```, este corresponde a la operación módulo, la cual entrega el resto de dividir el operando de la izquierda por el de la derecha. Una referencia extensa de los operadores disponibles en python se puede encontrar [aquí](https://j2logo.com/python/tutorial/operadores-en-python/).

Retornar multiples valores desde las funciones

Algunas alternativas son:

1. Separar los distintos valores usando comas (útil cuando cuando no se retornan demasiados valores)

2. Empaquetando los resultados en listas

3. Usando la estructura de datos diccionarios (lo veremos más adelante)

4. Usando tuplas

````python
# Forma 1
def operate(a, b):
    sum = a + b
    diff = a - b
    mul = a * b
    div = a / b
    return sum, diff, mul, div
````

````python
n1 = 5
n2 = 10
sum, diff, mul, div = operate(n1, n2)
print(
    f"La suma es {sum}\n"
    f"La diferencia es {diff}\n"
    f"La multiplicación da {mul}\n"
    f"La división da {div}\n"
)
````

````python
# si llamamos la función anterior y la asignamos a una sola variable, esta será de tipo tupla
resultado = operate(n1,n2)
print(resultado, type(resultado))
````

Función map():

Esta función devuelve un objeto mapeado de los resultados luego de aplicar la función dada a cada uno de los elementos de un objeto iterable (ej. lista, tupla, etc). Su sintaxis es `map(fun,iter)`

````python
#Ejemplo 1
#definamos una función que entregue el doble de un número n
def doblan(n):
    return 2*n

# luego, podemos duplicar todos los número de una lista usando map
numbers = [1,2,3,4]
result  = map(doblan,numbers)
print(list(result))
````

````python
# Ejemplo 2
# también se puede implementar con funciones anónimas
numbers = (5, 6, 7, 8)
result = map(lambda x: x + x, numbers)
print(list(result))
````
